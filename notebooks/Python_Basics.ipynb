# 1. What programming actually is?

### Reality of a computer is that it's not “smart.”
### It is just:

### Memory → places that hold values

### A processor → blindly changes those values

### Time → instructions executed one after another

### At every moment, the entire machine is just in one giant state 
### Memory + registers + flags + I/O = current state

### Programming is designing how that state changes over time or a program is just a story of how the state of memory changes over time.

### You are not telling the computer what something is, you are telling it how the world should evolve step by step.

### Every program is essentially:

### state₀ → state₁ → state₂ → state₃ → ...

### Your job is to shape these transitions intentionally

# 3. What is an Object?

### An object is simply: A bundle of data + behavior, living at some location in memory.

### More precisely: An object has:

### Identity – where it exists in memory

### State – the data it holds

### Behavior – what operations can be performed on it

### Think of it like a real-world thing.

### Example: A pen

### Identity → this specific pen in your hand

### State → color = blue, ink_level = 40%

### Behavior → write(), leak(), break()


## it's a later concept but good to introduce it here because it's not a concept which is language dependent 
## A class is a blueprint
## An object is memory shaped by that blueprint
## A variable is just a name pointing to that memory

# 5. After understanding how variable works and how everything in python is an object the right topic to move towards is mutability and immutability

### NOTE: if you are not familiar with data structures in any language then you can think for now that string, list, tuple, dictionary, set are just different type of data storing unit which store data in differnt ways 

## Once you understand that variables only point to objects, mutability becomes the question of whether the object itself can change not whether the variable can.

## You already know:

### Variables don’t store values,They point to objects

### Now the next question becomes:

### “When I use a variable, am I changing the object it points to — or just pointing it somewhere else?”

### That’s mutability vs immutability.

## Two kinds of objects
### Immutable objects
###   Cannot change after creation
###    Examples: int, float, str, tuple
### x = 5
### x += 1

### What actually happens:

### 5 is untouched

### a new object 6 is created

### x now points to it



```python
x = 5
x += 1
x
```

## Mutable objects

### Can change internally

### Examples: list, dict, set

### a = [1, 2]
### a.append(3)


### The object itself changes.

# 6. Import in Python: 
### In Python, import means load another file (module) and give access to its objects.

### For example:

### import math


### Python:

### Finds the math module

### Runs its code once

### Stores it as a module object

### Binds the name math to that object

### So:

### math.sqrt(16)


### means “use the sqrt function inside the math module”.

### Other forms
### from math import sqrt


### → brings sqrt directly into your namespace.

### import numpy as np


### → just renames the module for convenience.

### Key idea

### Import doesn’t copy code — it binds names to already-loaded objects.

```python
import math
```

```python
math.sqrt(16)
```

# 7. Copies -> (Assignments,Shallow,Deep)

# When you copy something, you’re asking:

## “Do I want a new object, or just another reference to the same object?”

### There are three cases in Python:

### Assignment (=)

### Shallow copy

### Deep copy

# Assignment (=) → No copy at all

```python
a = [1, 2, 3]
b = a
a
```

```python
b
```

### a ──┐
###     ├──► [1, 2, 3]
### b ──┘

# Shallow Copy → new container, same inner objects

```python

import copy 
a=[1,2,[3,4]]
b=copy.copy(a)
a
```

```python
b
```

### what' happens here is : 
### a ──► [ 1 , 2 , ──► [3, 4] ]
### b ──► [ 1 , 2 , ──► [3, 4] ]

### Outer list is new
### Inner list is shared

```python
# so 
b.append(5)      # only b changes
a
```

```python
b
```

```python
b[2].append(99)  # affects both
a
```

```python
b
```

# Deep Copy → completely independent

```python
import copy

a = [1, 2, [3, 4]]
b = copy.deepcopy(a)
```

```python
a
```

```python
b
```

### what it does
### a ──► [ 1 , 2 , [3, 4] ]
### b ──► [ 1 , 2 , [3, 4] ]


### Operation	What it does
### =	             new reference
### copy.copy()	     new outer object
### copy.deepcopy()	 full independent copy
